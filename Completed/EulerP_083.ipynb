```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import numpy as np
import pandas as pd

fiveByFive = np.array([
    [131,673,234,103,18],
    [201,96,342,965,150],
    [630,803,746,422,111],
    [537,699,497,121,956],
    [805,732,524,37,331]
])

eightyByEighty = pd.DataFrame(pd.read_csv('p081_matrix.txt',header=None)).values

distances = {}
nodes = []

matrix = eightyByEighty

for i, row in enumerate(matrix):
    for j, col in enumerate(row):
        # top left corner only
        if i == 0 and j == 0:
            distances.update({
                (i,j):{
                    (i+1,j): {'weight': matrix[i+1,j]}, # down
                    (i,j+1): {'weight': matrix[i,j+1]}, # right
                }
            })
        # top row only, not last column
        elif i == 0 and j != len(matrix)-1:
            distances.update({
                (i,j):{
                    (i+1,j): {'weight': matrix[i+1,j]}, # down
                    (i,j+1): {'weight': matrix[i,j+1]}, # right
                }
            })
        # not top or bottom row, not last column
        elif i != 0 and i!= len(matrix)-1 and j != len(matrix)-1 and j != 0:
            distances.update({
                (i,j):{
                    (i+1,j): {'weight': matrix[i+1,j]}, # down
                    (i,j+1): {'weight': matrix[i,j+1]}, # right
                    (i,j-1): {'weight': matrix[i,j-1]}, # left
                    (i-1,j): {'weight': matrix[i-1,j]}, # up
                }
            })
        # last column only, not last row or first row
        elif i != len(matrix)-1 and i != 0 and j == len(matrix)-1:
            distances.update({
                (i,j):{
                    (i+1,j): {'weight': matrix[i+1,j]}, # down
                    (i,j-1): {'weight': matrix[i,j-1]}, # left
                    (i-1,j): {'weight': matrix[i-1,j]}, # up
                }
            })
        # last row only, not last column
        elif i == len(matrix)-1 and j != 0 and j != len(matrix)-1:
            distances.update({
                (i,j):{
                    (i-1,j): {'weight': matrix[i-1,j]}, # up
                    (i,j+1): {'weight': matrix[i,j+1]}, # right
                    (i,j-1): {'weight': matrix[i,j-1]}, # left
                }
            })
        # first row, last column
        elif i == 0 and j == len(matrix)-1:
            distances.update({
                (i,j):{
                    (i+1,j): {'weight': matrix[i+1,j]}, # down
                    (i,j-1): {'weight': matrix[i,j-1]}, # left
                }
            })
        # bottom left
        elif i == len(matrix)-1 and j == 0:
            distances.update({
                (i,j):{
                    (i-1,j): {'weight': matrix[i-1,j]}, # up
                    (i,j+1): {'weight': matrix[i,j+1]}, # right
                }
            })
        elif i != 0 and i != len(matrix)-1 and j ==0:
            distances.update({
                (i,j):{
                    (i-1,j): {'weight': matrix[i-1,j]}, # up
                    (i,j+1): {'weight': matrix[i,j+1]}, # right
                    (i+1,j): {'weight': matrix[i+1,j]}, # down
                }
            })
            

G = nx.DiGraph(distances)

path = nx.dijkstra_path(G, (0,0), (len(matrix)-1,len(matrix)-1))

path_total = matrix[(0,0)]
for i in range(len(path)-1):
    path_total += G[path[i]][path[i+1]]['weight']

print path_total
```
