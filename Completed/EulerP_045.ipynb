# Problem 45 - Triangular, pentagonal, and hexagonal

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle	 	$T_n=n(n+1)/2$	 	

1, 3, 6, 10, 15, ...

Pentagonal	 	$P_n=n(3n−1)/2$

1, 5, 12, 22, 35, ...

Hexagonal	 	$H_n=n(2n−1)$	

1, 6, 15, 28, 45, ...


It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.

Find the next triangle number that is also pentagonal and hexagonal.

```python
def triangle(n):
    return n*(n+1)/2

def pentagonal(n):
    return n*(3*n-1)/2

def hexagonal(n):
    return n*(2*n-1)

triangle_n = set(map(triangle, range(1,10**5)))
pentagonal_n = set(map(pentagonal, range(1,10**5)))
hexagonal_n = set(map(hexagonal, range(1,10**5)))


print sorted(triangle_n & pentagonal_n & hexagonal_n)
```
